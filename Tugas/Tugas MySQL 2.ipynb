```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1.Report the account representative for each customer.
run("""
    SELECT customername, 
            CONCAT(contactLastName,' ', contactFirstName) AS Rep, 
            salesRepEmployeeNumber,
            amount
    FROM 
            customers
                INNER JOIN
            payments ON customers.customerNumber = payments.customerNumber
                
    """)
```

```python
#2. Report total payments for Atelier graphique.
run("""
SELECT 
    customername,
    SUM(amount) AS totalpayment
FROM
    customers
        INNER JOIN
    payments ON customers.customernumber = payments.customernumber
WHERE 
    customername = 'Atelier graphique'
    """)
```

```python
#8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.
run("""
SELECT
    b.customername,
    SUM(a.amount) AS total
FROM
    payments AS a
    INNER JOIN
    customers AS b ON a.customernumber = b.customernumber
GROUP BY
    customername
ORDER BY
    total DESC
    """)
```

## Many to Many Relationship

```python
#1. List Product Sold by Order Date
run("""
SELECT
    a.productname,
    b.quantityordered,
    c.orderdate
FROM
    products AS a
    INNER JOIN
    orderdetails AS b ON a.productcode = b.productcode
    INNER JOIN
    orders AS c ON b.ordernumber = c.ordernumber
    """)
```

```python
#2.List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
SELECT 
    a.productName,
    c.orderdate
FROM
    products AS a
    INNER JOIN orderdetails AS b ON a.productcode = b.productcode
    INNER JOIN orders AS c ON b.ordernumber = c.ordernumber
WHERE 
    productname = '1940 Ford Pickup Truck'
ORDER BY orderdate DESC
    """)
```

```python
#3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
SELECT 
    a.customername,
    b.ordernumber,
    b.orderdate,
    c.amount
FROM
    customers AS a
    INNER JOIN orders AS b ON a.customernumber = b.customernumber
    INNER JOIN payments AS c ON a.customernumber = c.customernumber 
WHERE amount > 25000
ORDER BY ordernumber
    """)
```

```python
#4.Are there any products that appear on all orders?
run("""
SELECT
    productname,
    ordernumber
FROM
    products
    RIGHT JOIN
    orderdetails ON products.productcode = orderdetails.productcode
WHERE productname = 'NaN'
ORDER BY ordernumber
    """)
```

```python
#5. List the names of products sold at less than 80% of the MSRP.
run("""
SELECT 
    a.productname,
    a.buyprice, 
    a.MSRP, 
    c.status,
    ((MSRP/buyprice)-1)*100 AS Markup_Percent
FROM products AS a
    INNER JOIN
    orderdetails AS b ON a.productcode = b.productcode
    INNER JOIN
    orders AS c ON b.ordernumber = c.ordernumber
WHERE ((MSRP/buyprice)-1) < 0.8 AND status = 'Shipped'
""")
```

```python
#6.Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
run("""
SELECT 
    a.productname,
    a.buyprice, 
    a.MSRP, 
    c.status,
    ((MSRP/buyprice)-1)*100 AS Markup_Percent
FROM products AS a
    INNER JOIN
    orderdetails AS b ON a.productcode = b.productcode
    INNER JOIN
    orders AS c ON b.ordernumber = c.ordernumber
WHERE ((MSRP/buyprice)-1) > 1 AND status = 'Shipped'
""")
```

```python
#7. List the products ordered on a Monday.
run("""
SELECT 
    productname,
    orderdate
FROM
    products AS a
    INNER JOIN
    orderdetails AS b ON a.productcode = b.productcode
    INNER JOIN
    orders AS c ON b.ordernumber = c.ordernumber
WHERE WEEKDAY(orderdate) = 0
    """)
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT 
    SUM(a.ordernumber) AS Hold_Number,
    b.status
FROM
    orderdetails AS a
    INNER JOIN
    orders AS b ON a.ordernumber = b.ordernumber
WHERE status = 'On Hold'
    """)
```
